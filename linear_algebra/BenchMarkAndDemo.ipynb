let's make some really really badly conditioned matrices.

```python
Vec = rand(1, 2000);
BadMatrix = Vec.T @ Vec
BadMatrix = BadMatrix + rand(BadMatrix.shape[0], BadMatrix.shape[1])*1e-2
plt.imshow(BadMatrix)
```

Let's try the LU Decomposition on this thing

```python
from linear_algebra.plu_decomposition import lu_decompose
P, L, U = lu_decompose(BadMatrix)
plt.imshow(L)
plt.show()
plt.imshow(U)
plt.show()
Reconstructed = P.T@L@U
print(f"Error Under 2 Norm: {norm(Reconstructed - BadMatrix)}")
plt.imshow(Reconstructed)
plt.show()



```
